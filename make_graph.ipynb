```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create dummy Tesla stock data
dates = pd.date_range(start="2023-01-01", periods=30, freq="D")  # 30 days of data
tesla_data = pd.DataFrame({
    "Date": dates,
    "Close": np.random.uniform(150, 300, size=30)  # Random stock closing prices
})

# Define the function to plot the graph
def make_graph(df, title):
    plt.figure(figsize=(10, 5))
    plt.plot(df["Date"], df["Close"], marker="o", linestyle="-", color="b")
    plt.xlabel("Date")
    plt.ylabel("Closing Price (USD)")
    plt.title(title)
    plt.xticks(rotation=45)
    plt.grid()
    plt.show()

# Call the function to display the graph
make_graph(tesla_data, "Tesla Stock Price Trend")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create dummy GameStop stock data
dates = pd.date_range(start="2023-01-01", periods=50, freq="D")  # 50 days of data
prices = np.random.randint(50, 300, size=50)  # Random stock prices between 50 and 300

# Convert to DataFrame
gme_data = pd.DataFrame({"Date": dates, "Stock Price": prices})

# Function to plot the graph
def make_graph(data, title):
    plt.figure(figsize=(10, 5))
    plt.plot(data["Date"], data["Stock Price"], marker="o", linestyle="-", color="b")
    plt.xlabel("Date")
    plt.ylabel("Stock Price (USD)")
    plt.title(title)
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.show()

# Plot GameStop stock data
make_graph(gme_data, "GameStop Stock Data")
```
