```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample dataset: Yearly sales for different vehicle types
data = {
    'Year': [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023],
    'SUV': [20000, 22000, 21000, 25000, 27000, 23000, 29000, 31000, 30000],
    'Sedan': [15000, 14000, 13500, 13000, 12500, 10000, 11000, 12000, 11500],
    'Truck': [18000, 19000, 18500, 20000, 21000, 18000, 22000, 23000, 22500]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Set 'Year' as index for easy plotting
df.set_index('Year', inplace=True)

# Define recession periods (Example: 2020)
recession_years = [2020]

# Plot the data
plt.figure(figsize=(10, 6))
for column in df.columns:
    plt.plot(df.index, df[column], marker='o', linestyle='-', label=column)

# Highlight recession periods with a shaded area
for year in recession_years:
    plt.axvspan(year - 0.5, year + 0.5, color='gray', alpha=0.3, label="Recession" if year == recession_years[0] else "")

# Customize the plot
plt.title("Vehicle Sales Trends by Type (2015-2023)")
plt.xlabel("Year")
plt.ylabel("Number of Sales")
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

def create_title_image(title, filename='Title.png'):
    plt.figure(figsize=(10, 2))
    plt.text(0.5, 0.5, title, fontsize=24, fontweight='bold', ha='center', va='center')
    plt.axis('off')  # Hide axes
    plt.savefig(filename, dpi=300, bbox_inches='tight')
    plt.show()

# Generate title image
title_text = "Automobile Sales Analysis Dashboard"
create_title_image(title_text)
```

```python
from PIL import Image, ImageDraw, ImageFont

# Define image size
img_width = 800
img_height = 200
background_color = (30, 30, 30)  # Dark background
text_color = (255, 255, 255)  # White text

# Create an image
img = Image.new("RGB", (img_width, img_height), background_color)
draw = ImageDraw.Draw(img)

# Define the title text
title_text = "Automobile Sales Analysis Dashboard"

# Load font (Change the path if using a custom font)
try:
    font = ImageFont.truetype("arial.ttf", 40)
except IOError:
    font = ImageFont.load_default()

# Get text size
text_bbox = font.getbbox(title_text)
text_width = text_bbox[2] - text_bbox[0]
text_height = text_bbox[3] - text_bbox[1]

# Calculate text position
x = (img_width - text_width) // 2
y = (img_height - text_height) // 2

# Add text to image
draw.text((x, y), title_text, fill=text_color, font=font)

# Save the image
img.save("Title.png")

# Show the image (optional)
img.show()
```

```python
import tkinter as tk
from tkinter import ttk
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg

# Callback Function to Update Output
def update_output(*args):
    selected_vehicle = vehicle_var.get()
    selected_period = period_var.get()
    output_label.config(text=f"Selected: {selected_vehicle} during {selected_period}")

# Callback Function to Update Graph
def update_plot():
    selected_vehicle = vehicle_var.get()
    selected_period = period_var.get()
    
    years = list(range(2010, 2023))
    sales = [500 + i * 10 if selected_period == "Pre-Recession" else 600 - i * 15 for i in range(len(years))]  # Dummy Data
    
    ax.clear()
    ax.plot(years, sales, marker='o', linestyle='-', label=selected_vehicle)
    ax.set_title(f"Sales Trend for {selected_vehicle} ({selected_period})")
    ax.set_xlabel("Year")
    ax.set_ylabel("Sales Volume")
    ax.legend()
    canvas.draw()

    update_output()

# Create Main Window
root = tk.Tk()
root.title("Sales Analysis")

# Dropdown for Vehicle Type
tk.Label(root, text="Vehicle Type:").pack()
vehicle_var = tk.StringVar(value="Sedan")
vehicle_dropdown = ttk.Combobox(root, textvariable=vehicle_var, values=["Sedan", "SUV", "Truck"])
vehicle_dropdown.pack()
vehicle_var.trace_add("write", update_output)  # Call update_output when selection changes

# Dropdown for Time Period
tk.Label(root, text="Time Period:").pack()
period_var = tk.StringVar(value="Pre-Recession")
period_dropdown = ttk.Combobox(root, textvariable=period_var, values=["Pre-Recession", "Recession", "Post-Recession"])
period_dropdown.pack()
period_var.trace_add("write", update_output)  # Call update_output when selection changes

# Button to Update Graph
tk.Button(root, text="Update Graph", command=update_plot).pack()

# Output Display
output_label = tk.Label(root, text="Select options and click 'Update Graph'", font=("Arial", 12), fg="blue")
output_label.pack()

# Matplotlib Figure for Graph
fig, ax = plt.subplots(figsize=(5, 3))
canvas = FigureCanvasTkAgg(fig, master=root)
canvas.get_tk_widget().pack()

# Run GUI
root.mainloop()
```
